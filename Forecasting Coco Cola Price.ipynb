```python
df['quarter'] = 0
for i in range(42):
    p=df['Quarter'][i]
    df['quarter'][i]=p[0:2]
```

```python
df_dummies=pd.DataFrame(pd.get_dummies(df['quarter']),columns=['Q1','Q2','Q3','Q4'])
```

```python
data=pd.concat([df,df_dummies],axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['quarter']=le.fit_transform(data['quarter'])
```

```python
data.head()
```

```python
x=data.drop(['Sales','Quarter'],axis=1)
x.shape
```

```python
y=data['Sales']
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
x=sd.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
print("R2_score:",r2_score(y_pred,y_test)*100)
print("Mean_Squared_Error:",mean_squared_error(y_pred,y_test))
print("Mean_Absolute_Error:",mean_absolute_error(y_pred,y_test))
print("Root__Mean_Squared_Error:",np.sqrt(mean_squared_error(y_pred,y_test)))
```

```python
#PolynomialModel

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
poly_reg_model = LinearRegression()
poly_reg_model.fit(x_train,y_train)
y_pred_poly = poly_reg_model.predict(x_test)
```

```python
print("Root__Mean_Squared_Error:",np.sqrt(mean_squared_error(y_pred_poly,y_test)))
```
